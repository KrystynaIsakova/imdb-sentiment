```python
import os
import pandas as pd
from openai import OpenAI

client = OpenAI(
    api_key=os.getenv("OPENAI_API_KEY")
)

# Load dataset
df = pd.read_csv("data/IMDB Dataset.csv")

# Take a small sample to avoid spending too much money
df = df.sample(10, random_state=42)


def predict_sentiment(review):
    response = client.responses.create(
        model="gpt-4o-mini",
        input=f"""
Classify the sentiment of the following movie review.

Return only one word:
positive
or
negative

Review:
{review}
"""
    )

    return response.output_text.strip().lower()


df["gpt_sentiment"] = df["review"].apply(predict_sentiment)

print(df[["sentiment", "gpt_sentiment"]].head())
```

```python
accuracy = (df["sentiment"] == df["gpt_sentiment"]).mean()
```
